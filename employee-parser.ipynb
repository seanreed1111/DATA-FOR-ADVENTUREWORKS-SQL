```python
DATABASE = {
        'NAME': 'jupyterdb',
        'USER': 'jupyteruser',
        'PASSWORD': 'password',
        'HOST': 'localhost',
        'PORT': 5432,
}
```

```python
# construct an engine connection string
engine_string = "postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}".format(
    user = DATABASE['USER'],
    password = DATABASE['PASSWORD'],
    host = DATABASE['HOST'],
    port = DATABASE['PORT'],
    database = DATABASE['NAME'],
)

# create sqlalchemy engine
engine = create_engine(engine_string)
```

```python
q = "CREATE TABLE IF NOT EXISTS mytable(id integer);"

try:
    pd.read_sql_query(q,engine)
except Exception as e:
    print(e)
```

```python
q2 = "SELECT * FROM mytable;"
pd.read_sql_query(q2,engine)
```

```python
q3 = "DROP TABLE IF EXISTS mytable;"
try:
    pd.read_sql_query(q3,engine)
except Exception as e:
    print(e)
```
